```python
%sql
--Revenue with 7-day moving average
WITH daily AS(
  SELECT event_date, SUM(revenue) AS daily_revenue
  FROM ecommerce.gold.products
  GROUP BY event_date
)
SELECT event_date, daily_revenue, AVG(daily_revenue) OVER (ORDER BY event_date ROWS BETWEEN 6 PRECEDING AND CURRENT ROW) AS moving_avg_seven
FROM daily
ORDER BY event_date;
```

```python
%sql
-- Conversion funnel
SELECT category_code,
SUM(views) as views,
SUM(purchases) as purchases,
ROUND(SUM(purchases)*100.0/SUM(views),2) as conversion_rate
FROM ecommerce.gold.products

GROUP BY category_code;
```

```python
%sql
--Customer tiers
SELECT
  CASE WHEN cnt >= 10 THEN 'VIP'
        WHEN cnt >= 5 THEN 'Loyal'
        ELSE 'Regular' END as tier,
  COUNT(*) as customers,
  AVG(total_spent) as avg_ltv
  FROM (
    SELECT user_id, COUNT(*) as cnt, SUM(price) as total_spent
    FROM ecommerce.silver.events
    WHERE event_type = 'purchase' GROUP BY user_id)
GROUP BY tier;
```
