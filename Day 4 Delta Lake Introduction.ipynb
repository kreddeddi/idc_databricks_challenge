Reading csv file to data frame

```python
events= spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)
events.show(10)
```

Convert CSV to delta format

```python
events.write.format("delta").mode("overwrite").save("/Volumes/workspace/ecommerce/delta_files")
```

Creating delta table

```python
events.write.format("delta").saveAsTable("events_table")
```

```python
%sql
select * from events_table limit 5
```

```python
spark.sql("""
          CREATE TABLE events_delta
          USING DELTA
          AS SELECT * FROM events_table
          """)
```
