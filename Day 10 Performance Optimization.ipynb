```python
# Explain query
spark.sql("SELECT * FROM ecommerce.silver.events WHERE event_type = 'purchase'").explain(True)
```

```python
# Partitioned table
spark.sql("""
CREATE TABLE IF NOT EXISTS ecommerce.silver.events_partitioned
USING DELTA
PARTITIONED BY (event_date, event_type)
AS
SELECT * FROM ecommerce.silver.events""")
```

```python
# Optimize
spark.sql("OPTIMIZE ecommerce.silver.events_partitioned ZORDER BY (user_id, product_id)")
```

```python
# Benchmark
import time
start = time.time()
spark.sql("SELECT * FROM ecommerce.silver.events_partitioned WHERE user_id= 12345").count()
print(f"Time taken: {time.time()-start:.2f}s")
```
